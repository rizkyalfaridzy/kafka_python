```python
from confluent_kafka import Consumer
import json
import ast
```

```python
# Create consumer
consumer = Consumer({
    'bootstrap.servers':'localhost',
    'group.id': 'G2',
    'auto.offset.reset': 'latest'
})
consumer.subscribe(['user'])
```

```python
# make list to save msg
saved_msg = []
```

```python
# Wait for the data to be available in the topic

while True:
    msg = consumer.poll(10)
    saved_msg.append(msg.value().decode('utf-8'))

    if msg is None:
        print("NO Message...cont...!!!")
    if msg.error():
        print("Consumer error: {}".format(msg.error()))
        continue

    print('Received message: {}'.format(msg.value().decode('utf-8')))
    print(saved_msg)
```

```python
# Check saved_msg content
saved_msg
```

```python
# Convert all message from string to dictionary
saved_msg_copy = saved_msg.copy()
for count in range(len(saved_msg)):
    saved_msg_copy[count] = ast.literal_eval(saved_msg_copy[count])
saved_msg_copy
```

```python
# Make The JSON structure
user_data_consumer = {
    'users': 
        saved_msg_copy
    }
user_data_consumer
```

```python
# Convert DIctionary to JSON
user_json_object = json.dumps(user_data_consumer, indent=4)
```

```python
# Save JSON file
with open("data lake/user.json", "w") as outfile:
    outfile.write(user_json_object)
```
